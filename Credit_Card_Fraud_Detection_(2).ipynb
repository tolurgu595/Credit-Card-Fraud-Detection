### Handling Outliers

The 'Transaction_amount' and 'Average Amount/transaction/day' columns have few outliers which may affect the performance of Machine Learning algorithms

```python
def handle_outliers_iqr(data, factor=1.5):
    # Step 1: Calculate the IQR
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    # Step 2: Determine the lower and upper bounds
    lower_bound = q1 - factor * iqr
    upper_bound = q3 + factor * iqr

    # Step 3: Replace outliers with median value
    median = np.median(data)
    handled_data = np.where((data < lower_bound) | (data > upper_bound), median, data)

    return handled_data
```

Handling outlier in the 'Average Amount/transaction/day' column

```python
df['Average Amount/transaction/day_outlier'] = handle_outliers_iqr(df['Average Amount/transaction/day'])
# df.head()
```

Handling outlier in the 'Transaction_amount' column

```python
df['Transaction_amount_Outlier'] = handle_outliers_iqr(df['Transaction_amount'])
```

## Feature Engineering and Contruction

Create a  new column  called 'Total Transaction Amount' by multiplying 'Transaction_amount' and 'Total Number of declines/day'

```python
df['Total_Transaction-Amount'] = df['Transaction_amount'] * df['Total Number of declines/day']
```

Create 'Chargeback_Amount_Ratio' which is the ratio of the average chargeback amount to the transaction amount by dividing the '6_month_avg_chbk_amt' column by the 'Transaction_amount' column.

```python
df['Chargeback_Amount_Ratio'] = df['6_month_avg_chbk_amt'] / df['Transaction_amount']
df['Chargeback_Amount_Ratio'].replace(np.inf, np.nan, inplace=True)  # Replace inf with NaN
```

```python
# Plot the histogram
df['Chargeback_Amount_Ratio_'] = df['Chargeback_Amount_Ratio'].fillna(df['Chargeback_Amount_Ratio'].mean())
```

## Binning

Binning 'Total Number of declines/day' into low, medium and high

```python
# Define the binning thresholds
thresholds = [-1, 5, 10, float('inf')]

# Define the bin labels
labels = ['low', 'medium', 'high']

# Create a new column with the bin labels
df['Declines_Level'] = pd.cut(df['Total Number of declines/day'], bins=thresholds, labels=labels)

# Print the resulting DataFrame
df.tail()
```

### Select the top best features using Univariate Selection

```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2, f_classif


# For numerical features (ANOVA F-value test)
numerical_features = X.select_dtypes(include=np.number)  # Select numerical features
selector_f = SelectKBest(score_func=f_classif, k='all')  # Choose the appropriate 'k' value
selector_f.fit(numerical_features, Y)
numerical_scores = pd.DataFrame({'Feature': numerical_features.columns, 'Score': selector_f.scores_})
numerical_scores = numerical_scores.sort_values('Score', ascending=False)

# Combine scores for all features
all_scores = pd.concat([numerical_scores])

# Print the top k features
k = 16  # Number of top features to select
top_k_features = all_scores.nlargest(k, 'Score')
print(top_k_features)
```
